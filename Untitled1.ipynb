```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import StackingClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
import lightgbm as lgb
import xgboost as XGBClassifier
```

```python
df_train=pd.read_csv('/content/train.csv')
df_test=pd.read_csv('/content/test.csv')
```

```python
df_train
#print(df_test.shape) # df_test has only 13 feature bcoz feature 'transported' has been removed 
# as it the target value which we have to predict.
#print(df_train.shape)
```

```python
categorical_features=['HomePlanet','Destination','Name']
# As we only train with categorical_features because in test person_id is given and we only
# need to predict whether he/she tranported or not.
float_features=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
# flost_features are collection of dtypes='int'
bool_features=['CryoSleep','VIP']
# bool _feature has true and false feature.
enc=LabelEncoder()
# LabelEncoder only assign numerical value for feastures in categorical_features,float_feature ,bool_feature
# just like ordinal Encoding.
# we take LabelEncoding in variable 'enc' to avoid again and again writting.
```

```python
def preprocessing(data,unk='unk'):
    
    data['PassengerId']=data['PassengerId'].apply(lambda x: x.split('_'))
    data['Passenger']=data['PassengerId'].apply(lambda x: int(x[0]))
    data['Id']=data['PassengerId'].apply(lambda x: int(x[1]))
    data.pop('PassengerId')
    # Creating new column passenger and id . using lambda function we make array with counting separate words 
    # when '_' is occured.
    
    ## NOTE- inplace=true = it makes the changes in the original data.
    for col in categorical_features:
        data[col].fillna(unk,inplace=True)
        data[col]=enc.fit_transform(data[col])
    # NOTE- 'transform function' only do Normalization with mean and standard deviation. But 'fit_transform'
    # function first transform the train data using mean and std and then fit using 'fit' function.
    for col in bool_features:
        data[col].fillna(False,inplace=True)
        data[col]=data[col].apply(int)
        
        
    data.Cabin.fillna('0/0/0',inplace=True)
    #preprocessing Cabin
    data.Cabin=data.Cabin.apply(lambda x: x.split('/'))
    data['path_1']=data['Cabin'].apply(lambda x:ord(x[0]))
    data['path_2']=data['Cabin'].apply(lambda x:int(x[1]))
    data['path_3']=data['Cabin'].apply(lambda x:ord(x[2]))
    data.pop('Cabin')
    # it basically create 3 more columna with 'path1','path2','path3' by makin array x by split('/').
    
    
    for col in float_features:
        data[col]=data[col].fillna(np.nanmedian(data[col].values))
    # it fill the empty item(as 'int' bcoz float_feature is made of interger feature) with meadian of
    # column.
        
    data['Total']=0    
        
    for col in ['HomePlanet', 'Destination']:
        data[f'{col}_count_encoding'] = data[col].map(data[col].value_counts())
    
    for col in float_features:
        data['Total'] =  data['Total']+data[col]
    # A new Column 'Total' is created that provide expendature or total cost for all float_feature.
        
    # Categorize the Total Spending
    data['Spend_Category'] = ''
    data.loc[data['Total'].between(0, 1, 'left'), 'Spend_Category'] = 'Zero_Spend'
    data.loc[data['Total'].between(1, 800, 'both'), 'Spend_Category'] = 'Under_800'
    data.loc[data['Total'].between(800, 1200, 'right'), 'Spend_Category'] = 'Median_1200'
    data.loc[data['Total'].between(1200, 2700, 'right'), 'Spend_Category'] = 'Upper_2700'
    data.loc[data['Total'].between(2700, 100000, 'right'), 'Spend_Category'] = 'Big_Spender'
    data['Spend_Category'] = data['Spend_Category'].astype('category').cat.codes.astype("int")  
    # A new column 'Spend_Category' is made that basically has some categorical feature and after
    # last line it gets ordinal Encoded.
    
    # Categorize the Ages
    data['Age_Category'] = ''
    data.loc[data['Age'].between(0, 18, 'both'), 'Age_Category'] = 'Under_18'
    data.loc[data['Age'].between(18, 40, 'right'), 'Age_Category'] = 'Adult'
    data.loc[data['Age'].between(40, 60, 'right'), 'Age_Category'] = 'Middle_Age'
    data.loc[data['Age'].between(60, 100, 'right'), 'Age_Category'] = 'Over_60'
    data['Age_Category'] = data['Age_Category'].astype('category').cat.codes.astype("int")
    # A new column 'Age_Category' is created with some categorical feature and then gets ordinal Encoded.
                                
    
    # FINAL NOTE= "MAKING NEW COLUMNS WITH EXISTING COLUMN IS REFRRED AS FEATURE ENGINEERING . AND IN THIS 
    # WE ALSO DO FEATURE SCALNG KNOWN AS (MAY BE z_SCORE_SCALING )".
    # Total_columns=14 + passenger + id + path1 + path2 + paht3 + spend_Category + Age_Category + Total
    # - Passenger_Id - Cabin + Home_count_encoding + destination_count_encoding = 22. 
    return data
```

```python
df_train_preprocessed=preprocessing(df_train)
```

```python
# WARNING/CAUTION= " WE CANNOT RE-RUN THE CELL BCOZ 1ST TIME WHEN WE RUN CELL THE 'Transported' column
# is deleted and if we try to re-run it then it gives error in 1 st line becoz 'Transported' column
# is already deleted."
# here 'transported' column is copy in y_train and 'tansported' column is deleted from df_train.
y_train=df_train['Transported'].apply(lambda x: int(x))
df_train.pop('Transported')
X_train=df_train
```

```python
# here we Ensemble the calssifiers in which we combile some different Models to increase the predictive 
# accuracy.
estimators=[('rf',RandomForestClassifier(n_estimators=700)),('catb',CatBoostClassifier()),('lgb',GradientBoostingClassifier())]
clf=StackingClassifier(estimators=estimators, final_estimator=SVC())
```

```python
clf.fit(X_train,y_train)
```
